## Kaggle 타이타닉 연습문제

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model

titanic_df = pd.read_csv('./data/titanic/train.csv')
titanic_df
```

```python
titanic_df.info()
refined_titanic_df = titanic_df[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]

print()
refined_titanic_df.info()
refined_titanic_df = refined_titanic_df.dropna(how='any')

print()
refined_titanic_df.info()
```

```python
x_data = refined_titanic_df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
x_data['Sex'] = x_data['Sex'].apply(lambda x: 0 if x == 'female' else 1)
x_data
```

```python
x_data
```

```python
x_data.shape
```

```python
x_data[:, 2].size
```

```python
age_min = np.min(x_data[:, 2])
age_max = np.max(x_data[:, 2])

fare_min = np.min(x_data[:, 5])
fare_max = np.max(x_data[:, 5])

x_data[:, 2] = (x_data[:, 2] - age_min)/(age_max - age_min)
x_data[:, 5] = (x_data[:, 5] - fare_min)/(fare_max - fare_min)

x_data
```

```python
t_data = refined_titanic_df['Survived']
t_data
```

```python
# x_data = x_data.values
t_data = t_data.values.reshape(-1, 1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

keras_model = Sequential()
keras_model.add(Flatten(input_shape=(6,)))
keras_model.add(Dense(128, activation='relu'))
keras_model.add(Dense(64, activation='relu'))
keras_model.add(Dense(32, activation='relu'))
keras_model.add(Dense(16, activation='relu'))
keras_model.add(Dense(1, activation='sigmoid'))

keras_model.compile(optimizer=Adam(learning_rate=0.005),
                    loss='binary_crossentropy',
                    metrics=['binary_accuracy'])

keras_model.fit(x_data,
                t_data,
                epochs=1000,
                verbose=1)
```
