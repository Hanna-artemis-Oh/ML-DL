```python
import numpy as np
import pandas as pd
```

```python
# 1. 기상 관측 이래, 서울의 최고 기온이 가장 높았던 날은 언제였고, 몇도인가요?
# 2. 역사적으로 일교차가 가장 큰 날짜는 몇년 몇월 몇일 인가요?
# 3. 1년 중 평균적으로 일교차가 가장 큰 날짜는 몇월 몇일 인가요?
# 4. 가장 덥다고 알려 대구보다 서울이 더 더운날이 가장 많은 연도는 언제인가요?
```

```python
df_seoul = pd.read_csv('./data/weather/seoul.csv', skiprows=7, encoding='cp949')
df_seoul
```

```python
df_daegu = pd.read_csv('./data/weather/daegu.csv', skiprows=7, encoding='cp949')
df_daegu
```

```python
# 1. 기상 관측 이래, 서울의 최고 기온이 가장 높았던 날은 언제였고, 몇도인가요?

df_seoul.sort_values(by='최고기온(℃)', ascending=False)[['날짜', '최고기온(℃)']].iloc[[0]]
```

```python
# 2. 역사적으로 일교차가 가장 큰 날짜는 몇년 몇월 몇일 인가요?

df_seoul['일교차'] = df_seoul['최고기온(℃)'] - df_seoul['최저기온(℃)']
df_seoul.sort_values(by='일교차', ascending=False)[['날짜', '일교차']].iloc[[0]]
```

```python
# 3. 1년 중 평균적으로 일교차가 가장 큰 날짜는 몇월 몇일 인가요?

df_seoul['날짜만'] = df_seoul['날짜'].str[5:]
grouped = df_seoul.groupby(['날짜만'])
average = grouped.mean()
average.sort_values(by='일교차', ascending=False).iloc[[0]].index.to_list()
```

```python
# 4. 가장 덥다고 알려진 대구보다 서울이 더 더운날이 가장 많은 연도는 언제인가요?

df = pd.merge(df_seoul, df_daegu, on='날짜', how='outer')
df.drop('일교차', axis=1, inplace=True)
df.drop('연도', axis=1, inplace=True)
df.drop('날짜만', axis=1, inplace=True)

df_hotter = df[df['평균기온(℃)_x'] > df['평균기온(℃)_y']]
df_hotter['연도'] = df_hotter['날짜'].str[:4]
grouped = df_hotter.groupby(['연도'])
counted = grouped.count()
hottest = counted.max()
counted.sort_values(by='날짜', ascending=False).iloc[[0]].index.to_list()
# hottest
```

```python
# 주요컬럼 설명
# manufacturer : 제조회사
# displ : 배기량
# cyl : 실린더 개수
# drv : 구동 방식
# hwy : 고속도로 연비
# class : 자동차 종류
# model : 자동차 모델명
# year : 생산연도
# trans : 변속기 종류
# cty : 도시 연비
# fl : 연료 종류
```

```python
df = pd.read_csv('./data/mpg/mpg.csv', skiprows=13)
df
```

```python
# 1. displ(배기량)이 4 이하인 자동차와 5 이상인 자동차 중 
#    어떤 자동차의 hwy(고속도로 연비)가 평균적으로 더 높은지 확인하세요.

displ_under = df[df['displ'] <= 4]['hwy'].mean()
displ_over = df[df['displ'] >= 5]['hwy'].mean()
displ_under, displ_over
```

```python
# 2. 자동차 제조 회사에 따라 도시 연비가 다른지 알아보려고 한다. 
#    "audi"와 "toyota" 중 어느 manufacturer(제조회사)의 cty(도시 연비)가 
#    평균적으로 더 높은지 확인하세요.

audi_cty = df[df['manufacturer'] == 'audi']['cty'].mean()
toyota_cty = df[df['manufacturer'] == 'toyota']['cty'].mean()
audi_cty, toyota_cty
```

```python
# 3. "chevrolet", "ford", "honda" 자동차의 고속도로 연비 평균을 알아보려고 한다. 
#    이 회사들의 데이터를 추출한 후 hwy(고속도로 연비) 평균을 구하세요.

df[df['manufacturer'].isin(['chevrolet', 'ford', 'honda'])]['hwy'].mean()
```

```python
# 4. "audi"에서 생산한 자동차 중에 어떤 자동차 모델의 hwy(고속도로 연비)가 
#    높은지 알아보려고 한다. 
#    "audi"에서 생산한 자동차 중 hwy가 1~5위에 해당하는 자동차의 데이터를 출력하세요.

df[df['manufacturer'] == 'audi'].sort_values(by='hwy', ascending=False).iloc[:5]
```

```python
# 5. mpg 데이터는 연비를 나타내는 변수가 2개입니다. 
#    두 변수를 각각 활용하는 대신 하나의 통합 연비 변수를 만들어 사용하려 합니다. 
#    평균 연비 변수는 두 연비(고속도로와 도시)의 평균을 이용합니다. 
#    회사별로 "suv" 자동차의 평균 연비를 구한후 내림차순으로 정렬한 후 1~5위까지 데이터를 출력하세요.

new_df = df
new_df['avg_fuel_efficiency'] = (new_df['hwy'] + new_df['cty']) / 2
suv_df = new_df[new_df['class'] == 'suv']
grouped = suv_df.groupby(['manufacturer'])
grouped.mean().sort_values(by='avg_fuel_efficiency', ascending=False).iloc[:5]
```

```python
# 6. mpg 데이터의 class는 "suv", "compact" 등 자동차의 특징에 따라 
#    일곱 종류로 분류한 변수입니다. 어떤 차종의 도시 연비가 높은지 비교하려 합니다. 
#    class별 cty 평균을 구하고 cty 평균이 높은 순으로 정렬해 출력하세요.

class_group = df.groupby(['class'])
avg_cty = class_group.mean().sort_values(by='cty', ascending=False)['cty']
avg_cty
```

```python
# 7. 어떤 회사 자동차의 hwy(고속도로 연비)가 가장 높은지 알아보려 합니다. 
#    hwy(고속도로 연비) 평균이 가장 높은 회사 세 곳을 출력하세요.

manufac_group = df.groupby(['manufacturer'])
high_hwy = manufac_group.mean().sort_values(by='hwy', ascending=False)['hwy'][:3]
high_hwy
```

```python
# 8. 어떤 회사에서 "compact" 차종을 가장 많이 생산하는지 알아보려고 합니다. 
#    각 회사별 "compact" 차종 수를 내림차순으로 정렬해 출력하세요.

compact_group = df[df['class'] == 'compact'].groupby(['manufacturer'])
compact_count = compact_group.count().sort_values(by='class', ascending=False)['class']
compact_count
```
